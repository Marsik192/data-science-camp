```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs 
import matplotlib.pyplot as plt
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, 
                           cluster_std = 0.6, random_state = 0)
colors = np.array([plt.cm.Spectral(val)
          for val in np.linspace(0, 1, len(set(y)))])
plt.figure(figsize=(8,6))

plt.scatter(X[:,0], X[:,1], c= colors[y], s= 25)
```

```python
from sklearn.cluster import AgglomerativeClustering
clf = AgglomerativeClustering(n_clusters= 5,linkage="complete")
predict = clf.fit_predict(X)
plt.figure(figsize=(8,8))
plt.scatter(X[:,0], X[:,1], c= colors[predict], marker= '.', s=100) 
plt.title ('Agglomerative Complete Clustering')
```

```python
clf_single = AgglomerativeClustering(n_clusters= 5,linkage="single")
predict_single = clf_single.fit_predict(X)
plt.figure(figsize=(8,8))
plt.scatter(X[:,0], X[:,1], c= colors[predict_single], marker= '.', s=100) 
plt.title ('Agglomerative Single Clustering')
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1.9, min_samples = 4).fit(X)
predict_dbscan = dbscan.fit_predict(X)
colors=  np.array(['green','grey', 'orange', 'brown', 'blue', 'yellow'])
colors= np.r_[colors, np.array(['black']*100)]

plt.figure(figsize=(8,8))
plt.scatter(X[:,0], X[:,1], c= colors[predict_dbscan], marker= '.', s=100) 
plt.title ('DBSCAN Clustering')
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

clf = KMeans(n_clusters = 5, random_state = 201)
elbow = KElbowVisualizer(clf)
elbow.fit(X)
elbow.show()
```
