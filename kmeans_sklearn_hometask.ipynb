```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
X, y= make_classification(n_samples=200,
                                    n_features=2,
                                    n_informative=2,
                                    n_redundant=0,
                                    random_state=0,
                                    n_clusters_per_class=1,
                                    class_sep = 0.8)
plt.figure()
plt.scatter(X[:,0], X[:,1])
#
#
# Мені не вийшло зробити домашку, зі своїм алгоритмом кластеризування, тому щоб було хоть щось
# Я зробив sklearn кластеризацію, затестив декілька алгоритмів, візуалізував cost function
#
#
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state = 2023).fit(X)
predict = kmeans.predict(X)

```

```python
plt.scatter(X[:, 0], X[:, 1], c=predict, s=40)
```

```python
centers =kmeans.cluster_centers_
plt.title('kmeans')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=70)
```

```python
#створимо ще декілька обєктів для тестування різних методів кластеризації
kmeans_2 = KMeans(n_clusters = 3,init= 'random',algorithm='full' ,random_state = 2023).fit(X)
predict_2 = kmeans_2.predict(X)
centers_2 =kmeans_2.cluster_centers_
plt.title('kmeans_2')
plt.scatter(X[:, 0], X[:, 1], c=predict_2, s=40)
plt.scatter(centers_2[:, 0], centers_2[:, 1], c='red', marker='x', s=170)
plt.tight_layout()
```
